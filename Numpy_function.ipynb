### Question 1 :

1.  Considering the simple mathematical formular

      $y$ = $x^3$ + $\sin$ $(x)$


* Using numpy, generate 20 random values between 0 and 1 for the values of  **x** i.e  $x$ $E$ $[0,1]$

* Write a function for the formular and generate the corresponding values of y for the generated x values above

* Visualize the plot of the given equation with the generated values 

```python
## Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#set seed(20) and generate 20 random values(0,1)
np.random.seed(10)
x = np.sort(np.random.random(20))
```

```python
def calculate(x):
    """function generates the coressponding value of y
    """
    cube_x = np.power(x,3)
    sin_x = np.sin(x)
    y = cube_x + sin_x
    
    return y
    
```

from the above formular,  $y$ = $x^3$ + $\sin$ $(x)$, The value of $y$ is :

```python
y = calculate(x)
y
```

### Visualization

```python
fig = plt.figure(figsize = (14, 8))
# Plot y = cube_x + sin(x)
y = calculate(x)
plt.plot(x, y, 'rs-', label = '$x^3$ + $\sin$ $(x)$')
# Add features to our figure
plt.legend()
plt.grid(True, linestyle =':', linewidth=3)
plt.xlim([0, 2])
plt.ylim([0, 2])
plt.title("Plot of  y = $x^3$ + $\sin$ $(x)$ ")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# Show plot
plt.show()
```
